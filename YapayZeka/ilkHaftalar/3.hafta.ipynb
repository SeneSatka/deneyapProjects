```python
import numpy as np
import pandas as pd
data =pd.read_csv("car.csv",encoding="unicode_escape")
data_copy=data.copy()

data_copy["fiyat"][len(data_copy["satis "])]=input("Fiyatı girin:")
data_copy["onarim"][len(data_copy["satis "])]=input("Onarımı girin:")
data_copy["kapi sayisi"][len(data_copy["satis "])]=input("Kapı sayısını girin:")
data_copy["kisi sayisi"][len(data_copy["satis "])]=input("Kişi sayısını girin:")
data_copy["bagaj boyutu"][len(data_copy["satis "])]=input("Bagaj boyutunu girin:")
data_copy["Guvenlik"][len(data_copy["satis "])]=input("Güvenlik girin:")


from sklearn import preprocessing
sayisallastirma=preprocessing.LabelEncoder()
data_copy["fiyat"]=sayisallastirma.fit_transform(data_copy["fiyat"])
data_copy["onarim"]=sayisallastirma.fit_transform(data_copy["onarim"])
data_copy["kapi sayisi"]=sayisallastirma.fit_transform(data_copy["kapi sayisi"])
data_copy["kisi sayisi"]=sayisallastirma.fit_transform(data_copy["kisi sayisi"])
data_copy["bagaj boyutu"]=sayisallastirma.fit_transform(data_copy["bagaj boyutu"])
data_copy["Guvenlik"]=sayisallastirma.fit_transform(data_copy["Guvenlik"])
data_copy["satis "]=sayisallastirma.fit_transform(data_copy["satis "])
inputs=np.array(data_copy.drop(["satis "],axis=1))
output=np.array(data_copy["satis "])
input_learn,satis_learn=inputs,output


from sklearn.naive_bayes import CategoricalNB
model = CategoricalNB()
model.fit(input_learn,satis_learn)
print(inputs[len(data_copy["satis "])-1])
satis_tahmin=model.predict([inputs[len(data_copy["satis "])-1]])
print(satis_tahmin)


# from sklearn.metrics import confusion_matrix,accuracy_score
# import seaborn as sns
# import matplotlib.pyplot as plt
# cm=confusion_matrix(satis_test,satis_tahmin)
# index=["Kolay","Normal","Zor","Çok kolay"]
# columns=["Kolay","Normal","Zor","Çok kolay"]
# cm_df=pd.DataFrame(cm,columns,index)
# plt.figure(figsize=(10,6))
# sns.heatmap(cm_df,annot=True,fmt="d",)
# print("Model doğruluğu: ",accuracy_score(satis_test,satis_tahmin))




```
